# Cálculo de Correlação


```python
# @title Lendo a planilha

import pandas as pd

df = pd.read_csv("../export/dados.csv", index_col=0)

ciclos = df.index.values
t_feed = df["t_feed"].values
t_rinse = df["t_rinse"].values
t_blow = df["t_blow"].values
t_purge = df["t_purge"].values

purity_H2 = df["purity_H2"].values
H2_CO_ratio = df["H2_CO_ratio"].values
purity_CO2 = df["purity_CO2"].values
recovery_CO2 = df["recovery_CO2"].values
productivity = df["productivity"].values
```

```python
inputs = [
    (t_feed, "$t_{feed}$"),
    (t_rinse, "$t_{rinse}$"),
    (t_blow, "t_{blow}$"),
    (t_purge, "$t_{purge}$"),
]

outputs = [
    (purity_H2, "H$_2$ purity"),
    (H2_CO_ratio, "H$_2$/CO ratio"),
    (purity_CO2, "CO$_2$ purity"),
    (recovery_CO2, "CO$_2$ recovery"),
    (productivity, "Productivity"),
]
```

## Autocorrelação


```python
from statsmodels.graphics.tsaplots import plot_acf

for output, output_name in outputs:
    plot_acf(output, title=f"Autocorrelation - {output_name}")
```

## Correlação Cruzada


```python
from statsmodels.graphics.tsaplots import plot_ccf

for input, input_name in inputs:
    for output, output_name in outputs:
        plot_ccf(
            input, output, title=f"Cross-correlation: {input_name} x {output_name}"
        )
```
