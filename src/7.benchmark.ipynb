# Benchmark


```python
# @title dados
import hickle as hkl
import numpy as np

arx = hkl.load("../export/benchmark_ARX.hkl")
# ss = hkl.load("../export/benchmark_state-space.hkl")
fnn = hkl.load("../export/benchmark_FNN.hkl")
rnn = hkl.load("../export/benchmark_RNN.hkl")

values = [
    ("ARX", arx * 1000),
    # ("State-Space", ss * 1000),
    ("FNN", fnn * 1000),
    ("RNN", rnn * 1000),
]
```

```python
import seaborn as sns

from lib.plot import colors, plt

for i in range(len(values)):
    model_name, times = values[i]
    mean_value = np.mean(times)

    sns.histplot(times, kde=True, stat="density", bins=15, color=colors[i])
    plt.axvline(
        mean_value,
        linestyle="--",
        linewidth=1.5,
        label=f"{model_name}: {mean_value:.3f}ms",
        color=colors[i],
    )
    plt.legend()
    plt.xlabel("Tempo / ms")
    plt.ylabel("Densidade de Probabilidade")

plt.savefig("../figures/benchmark.png")
plt.close()
```
