# Tratamento dos dados


```python
# @title Lendo a planilha
import pandas as pd

# Carregar o arquivo
df = pd.read_csv("../gPROMS/simul_Cgi.csv", skiprows=7)

df = df.drop(columns=["Variable", "X"])

# Renomear colunas
df = df.rename(columns={"NOCOMP": "Component", "Value": "Cgi"})

pivot = df.pivot_table(index="TIME", columns="Component", values="Cgi")

# Transformar tempo em array
t = pivot.index.to_numpy()

# Extrair as variáveis
CO2 = pivot[1].to_numpy()  # NOCOMP = 1
CO = pivot[2].to_numpy()  # NOCOMP = 2
H2 = pivot[3].to_numpy()  # NOCOMP = 3

print("t:", t[:5])
print("CO2:", CO2[:5])
print("CO:", CO[:5])
print("H2:", H2[:5])
```

```python
# @title plotando gráfico
from lib.plot import plt

t_h = t / (60 * 60)

plt.plot(t_h, CO2, label="CO$_2$")
plt.plot(t_h, CO, label="CO")
plt.plot(t_h, H2, label="H$_2$")

plt.xlabel("Tempo (h)")
plt.ylabel("Concentração na fase gás / mol$\\cdot$m$^3$")
plt.xlim(0, 2)

plt.legend()

plt.savefig("../figures/cgi-by-time.png")
plt.close()
```
